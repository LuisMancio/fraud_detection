Why you don’t see Pearson correlation with isFraud

Class imbalance – Fraud datasets usually have <1% fraud cases.

When 99% are 0 and only 1% are 1, correlations get “diluted”.

Pearson correlation assumes a linear relationship, which is often not there in fraud.

Categorical nature – Features like type, nameOrig, nameDest aren’t continuous variables. Encoding them numerically doesn’t magically make them linearly related to fraud.

Non-linear patterns – Fraud is often detected by interactions or thresholds (e.g., "large transfer + new receiver"), not simple linear trends.

df_feature['sender_avg_amount'] = df_feature.groupby('nameOrig')['amount'].transform('mean')

# Receiver features
df_feature['receiver_txn_count'] = df_feature.groupby('nameDest')['nameDest'].transform('count')
df_feature['receiver_avg_amount'] = df_feature.groupby('nameDest')['amount'].transform('mean')


### ✅ Safe Features You Can Engineer

With these 5 engineered features, you’ll give the model **transaction logic consistency checks** + **behavioral history**, which usually boost fraud detection.


### ⚠️ Features to Avoid (Leakage Risk)

* Anything derived from **`newbalanceOrig` or `newbalanceDest`** *alone* (because those are post-transaction values).
* Using **future aggregates** like “total amount sent in the next week” (not known at prediction time).
* Using `isFlaggedFraud` (it’s a regulatory marker, not part of natural features).



⚠️ Create a feature named diff_receiver_balance from old & new balance of destination/receiver

#### To check the balance difference after transaction on receiver accounts
 df_feature['diff_receiver_balance'] = df_feature['newbalanceDest'] - df_feature['oldbalanceDest']

# Transaction to Balance ratio

## Captures whether someone is sending a suspiciously large share of their balance.

```python
df_feature['txn_to_balance_ratio'] = df_feature['amount'] / (df_feature['oldbalanceOrg'] + 1)
```

⚠️ **Sender Balance Change (Expected vs Actual)**
   * For a *valid transaction*, this should be close to **0**.
   * If it deviates, that’s a fraud signal (many fraud records break this consistency).


df_feature['sender_balance_change'] = df_feature['oldbalanceOrg'] - df_feature['newbalanceOrig'] - df_feature['amount']

⚠️ **Receiver Balance Change (Expected vs Actual)**
 * Same logic → checks whether the receiver’s balance updated correctly.


df_feature['receiver_balance_change'] = df_feature['newbalanceDest'] - df_feature['oldbalanceDest'] - df_feature['amount']


**Transaction Frequency of Sender**
  * Counts how many transactions a sender has made so far.
  * Fraudsters often appear with **low history but sudden big transfers**.


```python
df_feature['txn_index'] = df_feature.groupby('nameOrig').cumcount() + 1
```

**Transaction Frequency of Receiver**
  * Some fraudulent receivers (money mule accounts) get *lots of incoming transactions*.
   * Helps flag high-traffic suspicious accounts.


```python
df_feature['receiver_txn_count'] = df_feature.groupby('nameDest').cumcount() + 1
```

**Time Since Last Transaction (`time_since_last_txn`)**

* Difference in `step` between current and previous transaction of the sender.
* **Why useful:** Fraud accounts may have **short gaps** between transactions (burst activity) or unusually **long inactivity** followed by sudden high-value transactions.


```python
df_feature['time_since_last_txn'] = df_feature.groupby('nameOrig')['step'].diff().fillna(0)
```

**Transactions per Hour (`txn_per_hour`)**

* Cumulative transaction count divided by elapsed hours.
* **Why useful:** Measures transaction velocity; very high or very low velocity can be suspicious.


```python
df_feature['txn_per_hour'] = df_feature['txn_index'] / (df_feature['step'] + 1)  # avoid division by 0
```

**Is First Transaction Flag (`is_first_txn`)**
* Marks whether this is the sender’s first transaction.
**Why useful:** Many fraud cases occur in the first transaction of a new or fake account.


```python
df_feature['is_first_txn'] = (df_feature['txn_index'] == 1).astype(int)
```

**Cumulative Amount for Sender (`cum_amount_sender`)**
* Total amount sent so far by this sender **before current transaction**.
* **Why useful:** Detects sudden spikes in transaction amounts.


Why shift(1) is important

Without shift(1), the cumulative sum includes the current transaction, which gives the model information about the present transaction, leading to data leakage.

Using shift(1) ensures that cum_amount_sender only uses past transactions to predict fraud for the current one.

```python
df_feature['cum_amount_sender'] = df_feature.groupby('nameOrig')['amount'].cumsum().shift(1).fillna(0)
```

1. Fraud detection: which is better?

transform('count') (global counts)

Tells you how “busy” an account is overall.

Example: if a sender usually only does 1–2 transactions, but suddenly has 50, it might be suspicious.

Problem: this uses all data, including “future” transactions → could cause data leakage if not careful.

cumcount() (running index)

Tells you which transaction number this is for that account.

Example: fraud is often more likely in the first transaction of a new account (fraudsters create fake accounts and hit once or twice, then disappear).

This avoids leakage because it only depends on the past order, not the entire dataset.

So cumcount() is safer and often more useful for fraud models.

✅ **Notes:**

* All features above are **leakage-free** — they only use **past and present information**.
* You can create similar **receiver-based features** using `nameDest`.
* Combining these with basic ratios (like `amount / oldbalanceOrg`) gives a strong fraud feature set.



#### We removed nameDest and nameOrig due to the high uniqueness of their values. Additionally, we will drop oldbalanceOrg, newbalanceOrig, oldbalanceDest, and newbalanceDest, as these features represent account balances after the transaction. Using them could introduce data leakage, causing the model to appear more accurate than it would be in a real-world scenario. In practice, fraud detection aims to identify suspicious transactions before they are completed, not after.
